```python
import json
import gradio as gr
from jsonschema import validate, ValidationError

# Company JSON Schema (Standard Format)
schema = {
    "type": "object",
    "properties": {
        "name": {"type": "string"},
        "email": {"type": "string"},
        "age": {"type": "number"}
    },
    "required": ["name", "email", "age"]
}


# JSON Validation Function
def validate_json(json_text):

    try:
        # Convert text to JSON
        data = json.loads(json_text)

        # Example Transformation (optional)
        data["name"] = data["name"].title()

        # Validate JSON against schema
        validate(instance=data, schema=schema)

        # Beautify JSON output
        refined = json.dumps(data, indent=4)

        return f"Validation Successful ✅\n\nRefined JSON:\n{refined}"

    except ValidationError as e:
        return f"Validation Error ❌\n\n{e.message}"

    except json.JSONDecodeError as e:
        return f"Invalid JSON Format ❌\n\n{str(e)}"


# Gradio Interface
interface = gr.Interface(
    fn=validate_json,
    inputs=gr.Textbox(lines=15, label="Paste Raw JSON Data"),
    outputs=gr.Textbox(lines=15, label="Validation Result"),
    title="JSON Refiner Advanced - QA Validation System",
    description="Validate JSON data against company schema and ensure data integrity"
)

interface.launch()
```
